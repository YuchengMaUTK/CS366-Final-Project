```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("web vulnerabilities.csv")
print(len(data))
#print(data)
data['date']= pd.to_datetime(data['date'], format='%Y-%m-%d')

data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

#print(data['year'].value_counts())

y9 = data.loc[(data['year']==2019)]
y8 = data.loc[(data['year']==2018)]
y7 = data.loc[(data['year']==2017)]
y6 = data.loc[(data['year']==2016)]

years = [y6,y7,y8,y9]

numT = []
for y in years:
    for i in range(1,13):
        m = y.loc[(y['month']==i)]
        numT.append(len(m))
        
print(numT)

```
